# Simple matshow() example

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
FIGSIZE = (10, 10)
FONT = {'family': 'sans-serif', 'weight': 'normal', 'size': 16}
```

## Simplest way :

```python
plt.matshow(np.random.rand(6, 6))

plt.show()
```

```python
plt.matshow(np.random.rand(6, 24), cmap=plt.cm.gray)

plt.show()
```

## More customisation

```python
def plot_matshow(matrix, x_tick_labels, y_tick_labels):
    """
    Plots a graphical representation of a given transition matrix
    """

    fig, ax = plt.subplots(figsize=FIGSIZE)
    ax.set_frame_on(False)
    
    cax = ax.matshow(matrix, cmap=plt.get_cmap("summer"))
    
    fig.colorbar(cax, fraction=0.046, pad=0.04)
    
    
    plt.tick_params(
    axis='x',       
    which='both',   
    bottom=True,      
    top=False,  
    labeltop=False,
    labelbottom=True)
    plt.xticks(range(len(x_tick_labels)), x_tick_labels)
    plt.yticks(range(len(y_tick_labels)), y_tick_labels)
    
    plt.title('Matshow example', fontdict=FONT, y=1.0)
    plt.xlabel('x : X label', fontdict=FONT)
    plt.ylabel('y : Y label', fontdict=FONT)
    
    plt.show()
```

```python
size = 6

mat = np.random.rand(2*size, size)

xlabels = ['x' + str(i+1) for i in range(size)]
ylabels = ['y' + str(i+1) for i in range(2*size)]

plot_matshow(mat, xlabels, ylabels)
```
